### Sleep Duration

```python
# Calculate Sleep Duration by hour
sleep_duration = minute_sleep.groupby(['Hour_of_the_day', 'value']).size().unstack(fill_value=0)
sleep_duration.columns = ['awake', 'light_sleep', 'deep_sleep']
print("Sleep Duration:")
print(sleep_duration)
```

```python
# Calculate Sleep Duration by week days
sleep_duration_weekdays = minute_sleep.groupby(['Number_of_week', 'value']).size().unstack(fill_value=0)
sleep_duration_weekdays.columns = ['awake', 'light_sleep', 'deep_sleep']
print("Sleep Duration by Week Days:")
print(sleep_duration_weekdays)
```

```python
plt.figure(figsize=(20, 12))

# SUBPLOT 1: BY HOUR
plt.subplot(2,1,1)

# Calculate total duration for each sleep stage at each time point
total_awake = sleep_duration['awake'].cumsum()
total_light_sleep = sleep_duration['light_sleep'].cumsum()
total_deep_sleep = sleep_duration['deep_sleep'].cumsum()

# Plot the cumulative sum of each sleep stage
plt.plot(sleep_duration.index, total_awake, label='Awake')
plt.plot(sleep_duration.index, total_light_sleep, label='Light Sleep')
plt.plot(sleep_duration.index, total_deep_sleep, label='Deep Sleep')

# Add labels and title
plt.title('Cumulative Sleep Stages Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Duration of Sleep Stage (minutes)')

# SUBPLOT 2: BY WEEKDAYS
plt.subplot(2,1,2)
# Calculate total duration for each sleep stage at each time point
total_awake_weekdays = sleep_duration_weekdays['awake'].cumsum()
total_light_sleep_weekdays = sleep_duration_weekdays['light_sleep'].cumsum()
total_deep_sleep_weekdays = sleep_duration_weekdays['deep_sleep'].cumsum()

# Plot the cumulative sum of each sleep stage
plt.plot(sleep_duration_weekdays.index, total_awake_weekdays, label='Awake')
plt.plot(sleep_duration_weekdays.index, total_light_sleep_weekdays, label='Light Sleep')
plt.plot(sleep_duration_weekdays.index, total_deep_sleep_weekdays, label='Deep Sleep')

# Add labels and title
plt.title('Cumulative Sleep Stages Over Week Days')
plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.xlabel('Date')
plt.ylabel('Cumulative Duration of Sleep Stage (minutes)')

# Add legend
plt.legend()

# Show plot
plt.show()
```

### Sleep Efficiency

```python
# Calculate Sleep Efficiency By hour

total_time_in_bed = sleep_duration.sum(axis=1)
total_sleep_time = sleep_duration[['light_sleep', 'deep_sleep']].sum(axis=1)

sleep_efficiency_hour = (total_sleep_time / total_time_in_bed) * 100

print("\nSleep Efficiency By Hour:")
print(sleep_efficiency_hour)
```

```python
# Calculate Sleep Efficiency by weekdays

total_time_in_bed_weekdays = sleep_duration_weekdays.sum(axis=1)
total_sleep_time_weekdays = sleep_duration_weekdays[['light_sleep', 'deep_sleep']].sum(axis=1)

sleep_efficiency_weekdays = (total_sleep_time_weekdays / total_time_in_bed_weekdays) * 100

print("\nSleep Efficiency By Week Days:")
print(sleep_efficiency_weekdays)
```
